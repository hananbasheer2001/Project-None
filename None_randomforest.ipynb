```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import dtreeviz
from sklearn.metrics import confusion_matrix
```

```python
# Load the iris dataset
iris = load_iris()

# Create a random forest classifier
clf = RandomForestClassifier(n_estimators=100, criterion="entropy")
```

```python
# Train the classifier on the iris dataset
clf.fit(iris.data, iris.target)
```

```python
# Decision Tree which we will use as an example
decision_tree = clf.estimators_[0]
tree = decision_tree.tree_
```

```python
# Visualize the tree graph
viz = dtreeviz.model(decision_tree, iris.data, iris.target, feature_names=iris.feature_names)
viz
```

```python
viz.view(scale=1, show_node_labels=True, orientation="TD")
```
